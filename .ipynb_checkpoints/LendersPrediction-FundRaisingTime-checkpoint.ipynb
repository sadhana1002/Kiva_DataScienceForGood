```python
import pandas as pd
```

```python
loans = pd.read_csv('loans.csv')

loans.columns
```

```python
loans.columns = map(lambda x:x.lower(),loans.columns)
```

```python
import matplotlib.pyplot as plt
```

```python
loans['English_Description']  = loans['description_translated'].fillna(loans['description'])
```

```python
loans.isnull().sum()
```

```python
loans_min = loans.drop(['description_translated','image_id','description','loan_name','original_language','video_id','town_name','currency_exchange_coverage_rate','currency_policy','planned_expiration_time','partner_id','tags'],axis=1)

loans_min.head()
```

```python
loans_min.isnull().sum()
```

```python
loans_reduced = loans_min.dropna(how='any')

loans_reduced.shape
```

```python
def gender_count(ppl,gender):
    count = 0
    for p in ppl:
        p = p.replace(' ','')
        if(p == gender):
            count = count+1
            
    return count

def female_count(ppl):
    return gender_count(ppl,'female')

def male_count(ppl):
    return gender_count(ppl,'male')
            
loans_reduced['borrowers'] = [str(x).split(',') for x in loans_reduced['borrower_genders']]
loans_reduced['female_count'] = loans_reduced['borrowers'].apply(female_count)
loans_reduced['male_count'] = loans_reduced['borrowers'].apply(male_count)
```

```python
loans_reduced = loans_reduced.drop(['borrowers','borrower_names','borrower_genders','borrower_pictured','country_code'], axis=1)

loans_reduced.head()
```

```python
loans_reduced['shortage_fund'] = loans_reduced['loan_amount'] - loans_reduced['funded_amount']

loans_reduced.loc[:,('posted_time')] = pd.to_datetime(loans_reduced['posted_time'],format='%Y-%m-%d %H:%M:%S')
loans_reduced.loc[:,('disburse_time')] = pd.to_datetime(loans_reduced['disburse_time'],format='%Y-%m-%d %H:%M:%S')
loans_reduced.loc[:,('raised_time')] = pd.to_datetime(loans_reduced['raised_time'],format='%Y-%m-%d %H:%M:%S')

from datetime import datetime, timedelta
loans_reduced.loc[:,'posted_raised_hours'] = (loans_reduced['raised_time']-loans_reduced['posted_time']).astype('timedelta64[h]')
loans_reduced.loc[:,'disburse_raised_hours'] = (loans_reduced['raised_time']-loans_reduced['disburse_time']).astype('timedelta64[h]')
```

```python
loans_reduced_df = loans_reduced.drop(['funded_amount','posted_time','disburse_time','raised_time'],axis=1)
```

```python
loans_reduced_df.to_csv('loans_reduced_dataset.csv',encoding='utf-8')
```

```python
country_samples = country_samples.reset_index(drop=True)

country_samples.head()
```

```python
country_samples['raised_time_class'] = pd.cut(country_samples['posted_raised_hours'],bins=[0,24,168,360,720,7555], labels=['one_day','week','fifteen','month','over_month'])
```

```python
country_samples = country_samples.set_index('loan_id')
```

```python
X = country_samples.drop(['raised_time_class','posted_raised_hours'], axis=1)
```

```python
# Classification - Trying to understand how much time would it take to classify

from sklearn import preprocessing
X = country_samples.drop(['disburse_raised_hours','raised_time_class'], axis=1)
X_dummies = pd.get_dummies(X)
Y = country_samples['raised_time_class']

encoder = preprocessing.LabelEncoder()
encoder.fit(list(Y))
Y_encoded = encoder.transform(list(Y))

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_dummies,Y_encoded,test_size=0.33, random_state=42)

from sklearn.neighbors import KNeighborsClassifier
train_scores = []
test_scores = []

for n in range(1,20,2):
    knn = KNeighborsClassifier(n_neighbors=n,weights='distance')
    knn.fit(x_train,y_train)
    train_score = knn.score(x_train,y_train)
    test_score = knn.score(x_test,y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f'N - {n}, Train - {train_score}, Test - {test_score}')

from sklearn.ensemble import RandomForestClassifier
rf_class = RandomForestClassifier(n_estimators=200)

rf_class.fit(x_train,y_train)

rf_train = rf_class.score(x_train,y_train)
rf_test = rf_class.score(x_test,y_test)

print(f'Random Forest Classifier Train - {rf_train}, Test - {rf_test}')
```

```python
X = country_samples.drop(['disburse_raised_hours','raised_time_class'], axis=1)
X_dummies = pd.get_dummies(X)

Y = country_samples['disburse_raised_hours']

x_train, x_test, y_train, y_test = train_test_split(X_dummies,Y,test_size=0.33, random_state=42)

from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor(n_estimators=100)

rf_reg.fit(x_train,y_train)

rf_train = rf_reg.score(x_train,y_train)
rf_test = rf_reg.score(x_test,y_test)

```
