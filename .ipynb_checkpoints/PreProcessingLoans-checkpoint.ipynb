```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
country_list = pd.read_html('https://developers.google.com/public-data/docs/canonical/countries_csv')

country_geo = country_list[0]

country_geo.columns = country_geo.iloc[0]

country_geo = country_geo[1:]

country_geo = country_geo.rename(columns={'country':'country_code'})

country_geo.head()
```

```python
loans = pd.read_csv('loans.csv')
```

```python
loans.columns = loans.columns.map(lambda x: x.lower())
```

```python
loans['shortage_fund'] =  loans['loan_amount'] -loans['funded_amount']
```

```python
loans.head()
```

```python
loans['English_Description']  = loans['description_translated'].fillna(loans['description'])


loans[['original_language', 'description','description_translated','English_Description']].head()
```

```python
def gender_count(ppl,gender):
    count = 0
    for p in ppl:
        p = p.replace(' ','')
        if(p == gender):
            count = count+1
            
    return count

def female_count(ppl):
    return gender_count(ppl,'female')

def male_count(ppl):
    return gender_count(ppl,'male')
            
loans['borrowers'] = [str(x).split(',') for x in loans['borrower_genders']]
loans['female_count'] = loans['borrowers'].apply(female_count)
loans['male_count'] = loans['borrowers'].apply(male_count)
loans['borrower_count'] = loans['male_count'] +loans['female_count'] 
```

```python
exchange_rate_df = pd.read_csv('exchangerate.csv')

exchange_rate_df.columns = [['currency', 'name', 'units_per_usd', 'usd_per_unit']]

exchange_rate_df.head()
```

```python
loans_usd = loans.merge(exchange_rate_df,on='currency')
```

```python
del loans
```

```python
loans_usd['funded_amount_usd'] = loans_usd['funded_amount']*loans_usd['usd_per_unit']
loans_usd['loan_amount_usd'] = loans_usd['loan_amount']*loans_usd['usd_per_unit']

loans_usd[['loan_amount_usd','funded_amount_usd']].head()
```

```python
loans_usd.columns
```
