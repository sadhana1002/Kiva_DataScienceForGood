```python
import pandas as pd
```

```python
loans = pd.read_csv('clustered_loans.csv')

# Remove class created by K-Means (using as shortcut to use cleaned stratified sample dataset)
loans = loans.drop('class',axis=1)

loans.columns
```

```python
import matplotlib.pyplot as plt
```

```python
loans_in_range = loans[loans["num_lenders_total"] <= loans["num_lenders_total"].quantile(0.95)]
```

```python
cat_cols = ['activity_name', 'sector_name','country','repayment_interval']

for col in cat_cols:
    col_amt = loans_in_range[[col,'num_lenders_total']]
    plt.figure(figsize=(30,10))
    ax = sns.boxplot(data=col_amt, x=col, y="num_lenders_total")
    plt.xlabel(col)
    plt.ylabel('Lenders')
    plt.xticks(rotation=90)
    plt.title(f'Lenders_vs_{col}')
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

X_raw = loans_in_range.drop(['num_lenders_total','loan_id','code'],axis=1)
X_dummies = pd.get_dummies(X_raw)

Y = loans_in_range['num_lenders_total']

x_scaler = StandardScaler().fit(X_dummies)
```

```python
X = x_scaler.transform(X_dummies)
```

```python
print(f"X shape - {X.shape}; Y shape - {Y.shape}")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y)
```

```python
loans_in_range.columns
```

```python
from sklearn.neighbors import KNeighborsRegressor
train_scores = []
test_scores = []

for n in range(3,20,2):
    knn = KNeighborsRegressor(n_neighbors=n,weights='distance')
    knn.fit(x_train,y_train)
    train_score = knn.score(x_train,y_train)
    test_score = knn.score(x_test,y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f'N - {n}, Train - {train_score}, Test - {test_score}')
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor(n_estimators=100)

rf_reg.fit(x_train,y_train)

rf_train = rf_reg.score(x_train,y_train)
rf_test = rf_reg.score(x_test,y_test)

```

```python
print(f'Random Forest Regressor Train - {rf_train}, Test - {rf_test}')
```

```python
import numpy as np

rf_reg.feature_importances_

# We can sort the features by their importance
all_features = sorted(zip(rf_reg.feature_importances_, X_dummies.columns), reverse=True)

top_features = all_features[:15]
```
