# Kiva - mining the loans through K-Means Clustering

In this notebook, an attempt is taken to identify natural groups in the dataset of worldwide loans from Kiva (Microfunding organization). 

Data can be found in https://build.kiva.org/docs/data/snapshots

Here is the approach taken:

    1.Get the preprocessed data (processed in other notebooks)   
    2.Remove outliers of the loan_amount field    
    3.Get poverty and other indices of countries (csv)    
    4.Take stratified sampling of the data (stratified by loan sector)
    5.Encode(Get dummies) for all categorical variables 
    6.Scale the sample dataset
    7.Apply hierarchial clustering (with different number of clusters) for small subset of samples to get a feel for cluster size
    8.Make an educated guess for 'K' (number of clusters)
    9.Run K-means algorithm with that K selected (Try this step with PCA (2 or 3 components)
    10.Use the labels given by Kmeans to understand the nature of the cluster

## Get Country Standards & Poverty Index Data

```python
# Get

mpi = pd.read_csv('mpi.csv')

mpi.head()
```

```python
mpi.columns
```

```python
mpi_reduced = mpi[['Country','Value','(%)',' Education','Health','Living standards','(%).3']]


mpi_reduced.columns = ['country','mpi','population_in_mpi','Education','Health','Living standards','population_in_severe_mpi']

mpi_reduced.head()
```

```python
countries_in_poverty = loans_reduced_df.merge(mpi_reduced,on='country',how='inner')

countries_in_poverty.head()
```

```python
countries_in_poverty = countries_in_poverty.dropna(how='any')
```

```python
countries_in_poverty.columns
```

## Take stratified sampling based on Loan sector

```python
# Stratified Sampling by country and sectors

loans_samples = countries_in_poverty.groupby(['sector_name'], group_keys=False).apply(lambda x: x.sample(min(len(x), 1000)))

loans_samples = loans_samples.reset_index(drop=True)
```

```python
loans_samples.shape
```

```python
# 'loan_id'
# 'loan_use'
# 'currency'
# 'num_bulk_entries'
# 'distribution_model'
# 'English_Description' 
# 'disburse_raised_hours'

loans_samples_reduced = loans_samples[['loan_amount', 'status', 'activity_name', 'sector_name',
       'country', 'lender_term', 'num_lenders_total',
       'num_journal_entries', 'repayment_interval',
       'female_count',
       'male_count', 'shortage_fund', 'posted_raised_hours',
       'mpi', 'population_in_mpi', 'Education',
       'Health', 'Living standards', 'population_in_severe_mpi']]
```

```python
loans_samples_reduced.head()
```

```python
categorical_columns = ['status', 'activity_name', 'sector_name', 'country', 'repayment_interval','shortage_fund' ]

for col in categorical_columns:
    print(loans_samples_reduced[col].value_counts())

# remove status - rarely distributed
# shortage fund - no different values

loans_samples_explore = .drop(['status','shortage_fund'],axis=1)

loans_samples_explore.head()
```

```python
loans_samples_reduced.shape
```

## Encode Categorical 

```python
loans_samples_dummies = pd.get_dummies(loans_samples_reduced)
```

## Scaling all features

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

scaler.fit(loans_samples_dummies)

loans_samples_scaled = scaler.transform(loans_samples_dummies)



features = pd.DataFrame(loans_samples_scaled)

features.columns = loans_samples_dummies.columns

features.head()
```

```python
features = pd.DataFrame(loans_samples_dummies)

features.columns = loans_samples_dummies.columns

features.head()
```

```python
features.to_csv('features_for_clustering.csv',encoding='utf-8')
```

## Hierarchial /Agglomerative Clustering

```python
hierarchial_samples = features.sample(1500)

hierarchial_samples.describe()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
import time
import collections

print("Compute structured hierarchical clustering...")

for n in range(3,10):
    hierarchial_samples = features.sample(1500)
    st = time.time()
    ward = AgglomerativeClustering(n_clusters=n,linkage='complete').fit(hierarchial_samples)
    elapsed_time = time.time() - st
    label = ward.labels_
    print(f"N: {n}")
    print(f"Elapsed time: {elapsed_time}")
    print(f"Number of points: {label.size}")
    
    col_name = f'labels[{n}]'
    hierarchial_samples[col_name] = label
   
    print(collections.Counter(label))
```

## Principal Component Analysis

To basically help with visualization 

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3) #-dimensional PCA

transformed = pd.DataFrame(pca.fit_transform(features))

transformed
```

## K Means - 6 clusters

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=6, random_state=0).fit(transformed)

kmeans
```

```python
label = kmeans.labels_

collections.Counter(label)


transformed['y'] = label

transformed.head()
```

```python
import mpl_toolkits.mplot3d.axes3d as p3
```

```python
fig = plt.figure(figsize=(10,10))
ax = p3.Axes3D(fig)
ax.view_init(7, -80)

ax.scatter( transformed[transformed['y']==0][1],  transformed[transformed['y']==0][2], transformed[transformed['y']==0][0],  label='Class 1', c='red')
ax.scatter( transformed[transformed['y']==1][1],  transformed[transformed['y']==1][2], transformed[transformed['y']==1][0],  label='Class 2', c='blue')
ax.scatter( transformed[transformed['y']==2][1],  transformed[transformed['y']==2][2], transformed[transformed['y']==2][0],  label='Class 3', c='lightgreen')

plt.legend()
plt.show()
```
