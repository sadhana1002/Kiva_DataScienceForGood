```python
def describe_categorical(df):

    cluster_frequency = []

    for k in range(n_clusters):
        kth_cluster = df[df['class']==k]

        category_values = {
            str(k) : {}
        }

        for c in ['country','sector_name','activity_name','status']:
            cat_values = kth_cluster[c].value_counts().head(10)
            
            cat_dict = [{x[0]:x[1]} for x in cat_values.items()]
            
            category_values[str(k)].update({str(c):cat_dict})

        cluster_frequency.append(category_values)
        
    return cluster_frequency
```
