```python
import pandas as pd
```

```python
loans = pd.read_csv('cleaned_loans.csv')
```

```python
# Import SQL Alchemy
from sqlalchemy import create_engine

# Import and establish Base for which classes will be constructed
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()

# Import modules to declare columns and column data types
from sqlalchemy import Column, Integer, String, Float, BLOB, DateTime
```

```python
engine = create_engine('mysql://root:root@localhost:3306/kiva')
```

```python
# Create both the Surfer and Board tables within the database
Base.metadata.create_all(engine)

# To push the objects made and query the server we use a Session object
from sqlalchemy.orm import Session
session = Session(bind=engine)
```

```python
loans.head()
```

```python
loan_records = loans.to_dict(orient='records')
```

```python
class Loans(Base):
    __tablename__ = 'loans'
    __table_args__ = {'extend_existing': True} 
    
    loan_id = Column(Integer, primary_key=True)
    status  = Column(String(50))
    sector_name = Column(String(50))
    activity_name = Column(String(50))
    country_code = Column(String(3))
    country_name = Column(String(50))
    posted_time = Column(DateTime)
    disburse_time = Column(DateTime)
    raised_time = Column(DateTime)
    lender_term = Column(Integer)
    num_lenders_total = Column(Integer)
    repayment_interval = Column(String(50))
    shortage_fund = Column(Float)
    female_count = Column(Integer)
    male_count = Column(Integer)
    borrower_count = Column(Integer)
    funded_amount_usd = Column(Float)
    loan_amount_usd = Column(Float)
    
    
    
    def __init__(self,data):
        print(data['loan_id'])
        self.loan_id = data['loan_id']
        self.status = data['status']
        self.sector_name = data['sector_name']
        self.activity_name = data['activity_name']
        self.country_code = data['country_code']
        self.country_name = data['country_name']
        self.posted_time = data['posted_time']
        self.disburse_time = data['disburse_time']
        self.raised_time = data['raised_time']
        self.lender_term = data['lender_term']
        self.num_lenders_total = data['num_lenders_total']
        self.repayment_interval = data['repayment_interval']
        self.shortage_fund = data['shortage_fund']
        self.female_count = data['female_count']
        self.male_count = data['male_count']
        self.borrower_count = data['borrower_count']
        self.funded_amount_usd = data['funded_amount_usd']
        self.loan_amount_usd = data['loan_amount_usd']
    
```

```python
for n in range(0,len(loan_records)-1000,1000):
    loan_objects = [Loans(x) for x in loan_records[n:n+1000]]
    session.bulk_save_objects(loan_objects)
    session.commit()
```
